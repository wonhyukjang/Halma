```python
import re
import os
def terminal_State(board, player):
        correct = 0
        if player == 'B':
            for i in range(2):
                for j in range(5):
                    if board[i+14][j+11] == 'B':
                        correct += 1
            
            for i in range(4):
                if board[13][i+12] == 'B':
                    correct += 1
            for i in range(3):
                if board[12][i+13] == 'B':
                    correct += 1
            for i in range(2):
                if board[11][i+14] == 'B':
                    correct += 1
        
        elif player == 'W':
            for i in range(2):
                for j in range(5):
                    if board[i][j] == 'W':
                        correct += 1

            for i in range(4):
                if board[2][i] == 'W':
                    correct += 1
            for i in range(3):
                if board[3][i] == 'W':
                    correct += 1
            for i in range(2):
                if board[4][i] == 'W':
                    correct += 1
        if correct == 19:
            return True
        return False
    
def generateInput():
    f = open("output.txt",'r')
    arr = []


    with open('output.txt') as openfileobject:
        for line in openfileobject:
            temp = re.findall(r'\d+', line) 
            res = list(map(int, temp[:2]))         
            res2 = list(map(int, temp[2:4]))         
            if res not in arr:
                arr.append(res)
            if res2 not in arr:
                arr.append(res2)
                
    start = arr[0][::-1]
    finish = arr[-1][::-1]

    f_in = open("input.txt",'r')

    # Check the mode
    mode = f_in.readline().rstrip()

    # Check who start first
    color = f_in.readline().rstrip()
    time = f_in.readline().rstrip()

    # Read initial position of the game
    board = []
    move_From = []
    move_To = []

    for i in range(16):
        board.append([x for x in f_in.readline().rstrip()])
    print("Before Moving =============================================")
    print("Player: ", color[0])
    showBoard(board)

    print("=============================================================")


    if board[start[0]][start[1]] == color[0]:
        print("Let's move a piece from: ", start[0],",",start[1])
        board[start[0]][start[1]] = '.'
    if board[finish[0]][finish[1]] == '.':
        print("Let's move to: ", finish[0],",",finish[1])
        board[finish[0]][finish[1]] = color[0]
    print("After Moving =============================================")
    print("Player: ", color[0])    
    showBoard(board)

    print("=============================================================")


    if os.path.isfile('input.txt'):
        os.remove('input.txt')
    
    with open('input.txt', 'w') as filehandle:
        filehandle.write('GAME')
        filehandle.write('\n')
        if color[0] == 'B':
            filehandle.write('WHITE')
            filehandle.write('\n')
        else:
            filehandle.write('BLACK')
            filehandle.write('\n')        
        filehandle.write('30.0')
        filehandle.write('\n')
        for j in range(len(board)):
            for i in range(len(board[j])):
                filehandle.write('%s' % board[j][i])
            if j != len(board) - 1:
                filehandle.write('\n')

```
